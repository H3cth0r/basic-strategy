```python
# --- Standard Imports ---
import yfinance as yf
import numpy as np
import pandas as pd
import ta
import matplotlib.pyplot as plt
import lightgbm as lgb
from sklearn.model_selection import TimeSeriesSplit, train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, roc_curve, precision_recall_curve
import vectorbt as vbt
import shap
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings

warnings.filterwarnings('ignore', category=FutureWarning)
warnings.filterwarnings('ignore', category=UserWarning)
#warnings.filterwarnings('ignore', category=lgb.basic.LightGBMWarning)

# --- Default Strategy Configuration ---
DEFAULT_PERIOD = "15d"
DEFAULT_INTERVAL = "5m"
DEFAULT_INIT_CASH = 10000
DEFAULT_RISK_PER_TRADE_PERCENT = 2.0
DEFAULT_FEES = 0.001
DEFAULT_SLIPPAGE = 0.0005

# --- Default Feature Engineering Parameters ---
DEFAULT_RSI_PERIOD = 14; DEFAULT_STOCH_PERIOD = 14; DEFAULT_MACD_FAST = 12; DEFAULT_MACD_SLOW = 26; DEFAULT_MACD_SIGN = 9
DEFAULT_ADX_PERIOD = 14; DEFAULT_BB_PERIOD = 20; DEFAULT_BB_STD_DEV = 2; DEFAULT_ATR_PERIOD = 14; DEFAULT_ATR_PERIOD_LONG = 100
DEFAULT_EMA_FAST_PERIOD = 12; DEFAULT_EMA_SLOW_PERIOD = 26; DEFAULT_EMA_TREND_PERIOD = 100

# --- Default Target Definition Parameters ---
DEFAULT_FUTURE_PERIODS = 15
DEFAULT_MOVE_THRESHOLD_ATR_MULTIPLIER = 1.5

# --- Default Model & Signal Parameters ---
DEFAULT_OPTIMIZE_THRESHOLD_METRIC = 'precision'
# DEFAULT_MIN_PRECISION_TARGET = 0.60 # Made asset-specific

# --- Default Risk Management Parameters (ATR Multiples) ---
DEFAULT_SL_ATR_MULTIPLIER = 1.5
DEFAULT_TP_ATR_MULTIPLIER = 2.5

# --- Default Training Parameters ---
DEFAULT_N_ESTIMATORS = 800
DEFAULT_EARLY_STOPPING_ROUNDS = 50
DEFAULT_APPLY_CLASS_WEIGHT = True
DEFAULT_USE_SIMPLIFIED_FEATURES = True

# --- Asset-Specific Overrides ---
# Define overrides here. Use ticker symbol as the key.
ASSET_CONFIG = {
    "ETH-USD": {
        "lgbm_params": { # Keep params that worked
            'learning_rate': 0.01,
            'num_leaves': 50,
            'reg_alpha': 0.1,
            'reg_lambda': 0.1,
         },
        "min_precision_target": 0.60,
        "use_trend_filter": True,
    },
    "BTC-USD": {
        "lgbm_params": { # Try different params for BTC
            'learning_rate': 0.02, # Slightly higher LR
            'num_leaves': 60,      # Slightly more leaves
            'reg_alpha': 0.05,     # Lower regularization initially
            'reg_lambda': 0.05,
        },
        "min_precision_target": 0.58, # Slightly lower precision target if needed
        "use_trend_filter": False, # *** Disable trend filter for BTC test ***
    },
    "XRP-USD": {
        "lgbm_params": { # Increase regularization for XRP
            'learning_rate': 0.01,
            'num_leaves': 50,
            'reg_alpha': 0.8, # *** Increased L1 regularization ***
            'reg_lambda': 0.8, # *** Increased L2 regularization ***
        },
        "min_precision_target": 0.62, # Try slightly higher precision target
        "use_trend_filter": True,
    }
    # Add other tickers and their specific configs here
}

# --- Helper to get config value ---
def get_config(ticker, key, default_value):
    return ASSET_CONFIG.get(ticker, {}).get(key, default_value)


# --- Data Fetching (Unchanged) ---
def create_dataframe(ticker, period, interval):
    print(f"Fetching data for {ticker} | Period: {period} | Interval: {interval}")
    try:
        df = yf.Ticker(ticker).history(period=period, interval=interval)
        df.columns = df.columns.str.replace(' ', '_')
        if df.empty: print(f"Warning: No data returned for {ticker}."); return None
        if df.index.tz is None: df.index = df.index.tz_localize('UTC')
        elif df.index.tz != 'UTC': df.index = df.index.tz_convert('UTC')
        print(f"Data fetched successfully for {ticker}. Shape: {df.shape}")
        return df
    except Exception as e: print(f"Error fetching data for {ticker}: {e}"); return None

# --- Feature Engineering (Reads USE_SIMPLIFIED_FEATURES from defaults) ---
def add_features(df, ticker):
    # Using default parameters for feature calculation
    use_simplified = DEFAULT_USE_SIMPLIFIED_FEATURES # Use the global default for feature set simplicity
    print(f"Adding features for {ticker} (Simplified: {use_simplified})...")

    df["rsi"] = ta.momentum.RSIIndicator(df['Close'], window=DEFAULT_RSI_PERIOD).rsi()
    df["stoch_osc"] = ta.momentum.StochasticOscillator(df["High"], df["Low"], df["Close"], window=DEFAULT_STOCH_PERIOD).stoch()
    macd = ta.trend.MACD(df['Close'], window_fast=DEFAULT_MACD_FAST, window_slow=DEFAULT_MACD_SLOW, window_sign=DEFAULT_MACD_SIGN)
    df['macd'] = macd.macd(); df['macd_signal'] = macd.macd_signal(); df['macd_diff'] = macd.macd_diff()
    df['adx'] = ta.trend.ADXIndicator(df['High'], df['Low'], df['Close'], window=DEFAULT_ADX_PERIOD).adx()
    bollinger = ta.volatility.BollingerBands(df['Close'], window=DEFAULT_BB_PERIOD, window_dev=DEFAULT_BB_STD_DEV)
    df['bb_high'] = bollinger.bollinger_hband(); df['bb_low'] = bollinger.bollinger_lband(); df['bb_width'] = bollinger.bollinger_wband()
    df['atr'] = ta.volatility.AverageTrueRange(df['High'], df['Low'], df['Close'], window=DEFAULT_ATR_PERIOD).average_true_range()
    df['obv'] = ta.volume.OnBalanceVolumeIndicator(df['Close'], df['Volume']).on_balance_volume()
    df['ema_fast'] = ta.trend.EMAIndicator(df['Close'], window=DEFAULT_EMA_FAST_PERIOD).ema_indicator()
    df['ema_slow'] = ta.trend.EMAIndicator(df['Close'], window=DEFAULT_EMA_SLOW_PERIOD).ema_indicator()
    df['hour'] = df.index.hour; df['dayofweek'] = df.index.dayofweek
    df['return_5p'] = df['Close'].pct_change(periods=5)
    df['atr_long'] = ta.volatility.AverageTrueRange(df['High'], df['Low'], df['Close'], window=DEFAULT_ATR_PERIOD_LONG).average_true_range()
    df['atr_ratio'] = (df['atr'] / df['atr_long']).replace([np.inf, -np.inf], np.nan).fillna(1.0)

    # Trend filter EMA added regardless of USE_TREND_FILTER flag here,
    # filtering happens during backtest based on asset config
    df['ema_trend'] = ta.trend.EMAIndicator(df['Close'], window=DEFAULT_EMA_TREND_PERIOD).ema_indicator()

    if not use_simplified:
        print("Adding slope and ROC features...")
        df['ema_fast_slope'] = df['ema_fast'].diff(); df['ema_slow_slope'] = df['ema_slow'].diff()
        df['rsi_roc'] = df['rsi'].diff(); df['macd_diff_roc'] = df['macd_diff'].diff()

    print(f"Features added. Shape before dropna: {df.shape}")
    initial_len = len(df)
    df.drop(columns=['atr_long'], inplace=True, errors='ignore')
    df.dropna(inplace=True)
    print(f"Dropped {initial_len - len(df)} rows with NaNs. Final shape: {df.shape}")
    return df


# --- Target Definition (Unchanged) ---
def add_target_rolling(df, future_periods=DEFAULT_FUTURE_PERIODS, threshold_atr_mult=DEFAULT_MOVE_THRESHOLD_ATR_MULTIPLIER):
    print(f"Adding rolling target: {future_periods} periods lookahead, {threshold_atr_mult}*ATR threshold...")
    df['atr_threshold'] = df['atr'] * threshold_atr_mult
    df['future_high'] = df['High'].shift(-future_periods).rolling(window=future_periods, closed='left').max()
    df['future_low'] = df['Low'].shift(-future_periods).rolling(window=future_periods, closed='left').min()
    price_increase_target = df['Close'] + df['atr_threshold']
    hit_target_up = df['future_high'] >= price_increase_target
    df['target'] = 0
    df.loc[hit_target_up, 'target'] = 1
    df.drop(columns=['atr_threshold', 'future_high', 'future_low'], inplace=True)
    initial_len = len(df)
    df.dropna(subset=['target'], inplace=True)
    print(f"Target added. Dropped {initial_len - len(df)} rows due to lookahead. Final shape: {df.shape}")
    target_counts = df['target'].value_counts(normalize=True)
    print(f"Target value counts:\n{target_counts}")
    if len(target_counts) < 2: print("Warning: Target variable has only one class."); return df.iloc[0:0] # Return empty df
    return df


# --- Train/Validation/Test Split (Reads USE_SIMPLIFIED_FEATURES) ---
def train_val_test_split_chronological(df, val_size=0.15, test_size=0.15):
    print(f"Splitting data chronologically. Val size: {val_size*100:.1f}%, Test size: {test_size*100:.1f}%")
    total_size = len(df)
    test_split_idx = int(total_size * (1 - test_size))
    val_split_idx = int(total_size * (1 - test_size - val_size))
    train, val, test = df.iloc[:val_split_idx], df.iloc[val_split_idx:test_split_idx], df.iloc[test_split_idx:]
    print(f"Train set size: {len(train)} ({train.index.min()} to {train.index.max()})")
    print(f"Validation set size: {len(val)} ({val.index.min()} to {val.index.max()})")
    print(f"Test set size: {len(test)} ({test.index.min()} to {test.index.max()})")

    use_simplified = DEFAULT_USE_SIMPLIFIED_FEATURES
    base_features = [
        "rsi", "stoch_osc", "macd", "macd_signal", "macd_diff", "adx",
        "bb_high", "bb_low", "bb_width", "atr", "obv",
        "ema_fast", "ema_slow", "hour", "dayofweek", "return_5p", "atr_ratio", "ema_trend"] # Always include ema_trend
    if not use_simplified:
        extra_features = ["ema_fast_slope", "ema_slow_slope", "rsi_roc", "macd_diff_roc"]
        feature_cols = base_features + extra_features
    else:
        feature_cols = base_features

    feature_cols = [col for col in feature_cols if col in df.columns]
    print(f"Using {len(feature_cols)} features: {feature_cols}")

    X_train, y_train = train[feature_cols], train['target']
    X_val, y_val = val[feature_cols], val['target']
    X_test, y_test = test[feature_cols], test['target']
    test_data_full = test.copy()
    return X_train, X_val, X_test, y_train, y_val, y_test, test_data_full, feature_cols


# --- Plotting Utilities (Unchanged) ---
# plot_confusion_matrix_and_metrics, plot_shap_and_importance, plot_correlation_matrix remain the same
def plot_confusion_matrix_and_metrics(y_true, y_pred, y_proba, class_names=None, title='Confusion Matrix and Metrics'):
    if y_true.empty or len(y_pred) == 0 or len(y_proba) == 0: print("Cannot plot confusion matrix/metrics: Empty data."); return
    if not (len(y_true) == len(y_pred) == len(y_proba)): print(f"Warning: y_true/y_pred/y_proba lengths differ. Cannot plot metrics."); return
    if class_names is None: class_names = ['Down/Neutral (0)', 'Up (1)']
    cm = confusion_matrix(y_true, y_pred); accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, zero_division=0); recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    try: auc = roc_auc_score(y_true, y_proba) if len(np.unique(y_true)) > 1 else np.nan
    except ValueError as e: print(f"Could not calculate AUC: {e}."); auc = np.nan
    fig = make_subplots(rows=1, cols=3, column_widths=[0.35, 0.25, 0.4], specs=[[{'type': 'heatmap'}, {'type': 'table'}, {'type': 'scatter'}]], subplot_titles=('Confusion Matrix', 'Metrics', 'ROC Curve'))
    fig.add_trace(go.Heatmap(z=cm, x=class_names, y=class_names, colorscale='Blues', text=cm, texttemplate="%{text}", showscale=False, name=''), row=1, col=1)
    metrics_data = [['Accuracy', f'{accuracy:.4f}'], ['Precision', f'{precision:.4f}'], ['Recall', f'{recall:.4f}'], ['F1 Score', f'{f1:.4f}'], ['AUC', f'{auc:.4f}' if not np.isnan(auc) else 'N/A']]
    metrics_table = go.Table(header=dict(values=['Metric', 'Value'], fill_color='paleturquoise', align='left'), cells=dict(values=[[m[0] for m in metrics_data], [m[1] for m in metrics_data]], fill_color='lavender', align='left'))
    fig.add_trace(metrics_table, row=1, col=2)
    if not np.isnan(auc): fpr, tpr, _ = roc_curve(y_true, y_proba); fig.add_trace(go.Scatter(x=fpr, y=tpr, mode='lines', name=f'ROC Curve (AUC = {auc:.4f})', line=dict(color='blue')), row=1, col=3)
    else: fig.add_annotation(text="AUC not computable", xref="paper", yref="paper", x=0.5, y=0.5, showarrow=False, row=1, col=3)
    fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], mode='lines', name='Random Chance', line=dict(color='grey', dash='dash')), row=1, col=3)
    fig.update_layout(title_text=title, height=400, width=1000, showlegend=False)
    fig.update_xaxes(title_text='Predicted Label', row=1, col=1); fig.update_yaxes(title_text='True Label', autorange='reversed', row=1, col=1)
    fig.update_xaxes(title_text='False Positive Rate', range=[0,1], row=1, col=3); fig.update_yaxes(title_text='True Positive Rate', range=[0,1], row=1, col=3)
    fig.show()

def plot_shap_and_importance(model, X_train_df, X_test_df, feature_names, title_suffix=''):
    print("Calculating SHAP values for test set...")
    if not isinstance(X_test_df, pd.DataFrame): print("Error: X_test_df must be a pandas DataFrame for SHAP plotting."); return
    if not hasattr(model, 'predict_proba'): print("Model does not have predict_proba, skipping SHAP plots."); return # TreeExplainer needs predict_proba

    try:
        explainer = shap.TreeExplainer(model)
        if X_test_df.empty: print("Skipping SHAP plots: X_test_df is empty."); return
        if not hasattr(model, '_Booster') or model._Booster is None: print("Error: Model doesn't appear fitted. Cannot calculate SHAP."); return

        shap_values_list = explainer.shap_values(X_test_df) # Pass DataFrame directly

        if isinstance(shap_values_list, list) and len(shap_values_list) == 2: shap_values_positive_class = shap_values_list[1]
        else: print("Warning: SHAP values format unexpected."); shap_values_positive_class = shap_values_list

        if not hasattr(shap_values_positive_class, 'shape') or len(shap_values_positive_class.shape) != 2: print(f"Error: Unexpected SHAP values shape."); return
        if shap_values_positive_class.shape[0] != X_test_df.shape[0]: raise ValueError("Shape mismatch (rows): SHAP vs X_test")
        if shap_values_positive_class.shape[1] != X_test_df.shape[1]:
             if shap_values_positive_class.shape[1] == X_test_df.shape[1] + 1: shap_values_positive_class = shap_values_positive_class[:, :-1] # Handle base value
             else: raise ValueError(f"Shape mismatch (columns): SHAP ({shap_values_positive_class.shape[1]}) vs X_test ({X_test_df.shape[1]})")

        print("Generating SHAP summary plots...")
        plt.close('all')
        plt.figure(figsize=(10, max(6, len(X_test_df.columns) * 0.3)))
        shap.summary_plot(shap_values_positive_class, X_test_df, plot_type="bar", show=False); plt.title(f"Feature Impact {title_suffix}"); plt.tight_layout(); plt.show()
        plt.figure(figsize=(10, max(6, len(X_test_df.columns) * 0.3)))
        shap.summary_plot(shap_values_positive_class, X_test_df, show=False); plt.title(f"SHAP Distribution {title_suffix}"); plt.tight_layout(); plt.show()
        print("SHAP plots generated.")

    except Exception as e: print(f"Could not generate SHAP plots: {e}"); import traceback; traceback.print_exc()

    print("Generating LGBM feature importance plot...")
    try:
        if hasattr(model, 'feature_importances_') and model.feature_importances_ is not None:
             if len(feature_names) == len(model.feature_importances_):
                 importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': model.feature_importances_}).sort_values('Importance', ascending=True)
                 fig = go.Figure(go.Bar(x=importance_df['Importance'], y=importance_df['Feature'], orientation='h'))
                 fig.update_layout(title=f'LGBM Feature Importance {title_suffix}', height=max(400, len(importance_df) * 20)); fig.show()
                 print("LGBM feature importance plot generated.")
             else: print(f"Feature name length ({len(feature_names)}) mismatch with importance length ({len(model.feature_importances_)}).")
        else: print("Model does not have feature_importances_ attribute.")
    except Exception as e: print(f"Could not generate LGBM importance plot: {e}")


def plot_correlation_matrix(data, title='Feature Correlation Matrix'):
    numeric_data = data.select_dtypes(include=np.number);
    if numeric_data.empty: print("No numeric data for correlation matrix."); return
    corr = numeric_data.corr()
    fig = go.Figure(data=go.Heatmap(z=corr.values, x=corr.columns, y=corr.columns, colorscale='RdBu', zmin=-1, zmax=1, text=np.around(corr.values, decimals=2), texttemplate="%{text}", colorbar=dict(title='Correlation')))
    fig.update_layout(title=title, width=min(1200, 40 * len(corr.columns)), height=min(1000, 40 * len(corr.columns)), xaxis_tickangle=-45, margin=dict(l=100, r=50, b=150, t=100)); fig.show()

# --- Probability Threshold Optimization (Unchanged) ---
def find_optimal_threshold(model, X_val, y_val, metric='precision', min_precision=0.6):
    print(f"\nOptimizing probability threshold on validation set using '{metric}' metric (Min Precision Target: {min_precision})...")
    if X_val.empty or y_val.empty: print("Validation set is empty."); return None
    try: y_pred_proba_val = model.predict_proba(X_val)[:, 1]
    except Exception as e: print(f"Error predicting probabilities on validation set: {e}"); return None
    precision, recall, thresholds = precision_recall_curve(y_val, y_pred_proba_val)
    thresholds = np.append(thresholds, 1.0)
    valid_indices = np.where(precision >= min_precision)[0]
    if len(valid_indices) > 0:
        f1_scores_valid = (2 * precision[valid_indices] * recall[valid_indices]) / (precision[valid_indices] + recall[valid_indices])
        f1_scores_valid = np.nan_to_num(f1_scores_valid)
        best_idx_in_valid = np.argmax(f1_scores_valid)
        optimal_threshold = thresholds[valid_indices][best_idx_in_valid]
        achieved_precision = precision[valid_indices][best_idx_in_valid]; achieved_recall = recall[valid_indices][best_idx_in_valid]; achieved_f1 = f1_scores_valid[best_idx_in_valid]
        print(f"Threshold meeting Precision >= {min_precision}: {optimal_threshold:.4f}")
        print(f"  Validation Metrics at Threshold: Precision={achieved_precision:.4f}, Recall={achieved_recall:.4f}, F1={achieved_f1:.4f}")
        optimal_threshold = min(0.99, optimal_threshold); return optimal_threshold
    else:
        max_prec = 0
        if len(precision) > 0 : max_prec = np.max(precision[np.isfinite(precision)])
        print(f"FAILED: Minimum precision {min_precision} not met on validation set. Max validation precision: {max_prec:.4f}")
        return None


# --- Backtesting Function (Reads asset-specific use_trend_filter) ---
def run_backtest(df_test, ticker, init_cash, size, size_type, fees, slippage, sl_atr_mult, tp_atr_mult, optimized_proba_threshold, interval_str):
    print("\n--- Starting Backtest ---")
    if optimized_proba_threshold is None: print("Skipping backtest: Optimal threshold not found."); return None, None
    if df_test.empty: print("Cannot backtest: Test data is empty."); return None, None

    # Get asset-specific trend filter setting
    use_trend_filter = get_config(ticker, 'use_trend_filter', True) # Default to True if not specified

    required_cols = ['Close', 'pred_proba', 'atr'] + (['ema_trend'] if use_trend_filter else [])
    if not all(col in df_test.columns for col in required_cols): missing = [c for c in required_cols if c not in df_test.columns]; print(f"Missing required columns: {missing}."); return None, None

    entries = df_test['pred_proba'] > optimized_proba_threshold
    if use_trend_filter: print(f"Applying trend filter for {ticker}: Close > EMA({DEFAULT_EMA_TREND_PERIOD})"); trend_condition = df_test['Close'] > df_test['ema_trend']; entries = entries & trend_condition
    else: print(f"Trend filter DISABLED for {ticker}")

    exits = pd.Series(False, index=df_test.index)
    sl_points = df_test['atr'] * sl_atr_mult; tp_points = df_test['atr'] * tp_atr_mult
    valid_close = df_test['Close'] > 1e-6
    sl_pct = pd.Series(np.nan, index=df_test.index); tp_pct = pd.Series(np.nan, index=df_test.index)
    sl_pct[valid_close] = sl_points[valid_close] / df_test['Close'][valid_close]; tp_pct[valid_close] = tp_points[valid_close] / df_test['Close'][valid_close]
    sl_pct.fillna(0, inplace=True); tp_pct.fillna(0, inplace=True)
    sl_pct = sl_pct.clip(lower=1e-5); tp_pct = tp_pct.clip(lower=1e-5)

    try:
        interval_map = {'1m': '1min', '5m': '5min', '10m': '10min', '15m': '15min', '30m': '30min', '1h': 'H', '1d': 'D'}
        freq = interval_map.get(interval_str, interval_str)
        print(f"Backtest Params {ticker} | Freq: '{freq}' | Threshold: {optimized_proba_threshold:.4f} | SL: {sl_atr_mult}*ATR | TP: {tp_atr_mult}*ATR | TrendFilter: {use_trend_filter}")
        if not entries.any(): print("No entry signals generated. Skipping portfolio simulation."); return None, None

        portfolio = vbt.Portfolio.from_signals(close=df_test["Close"], entries=entries, exits=exits, freq=freq, init_cash=init_cash, size=size, size_type='percent', fees=fees, slippage=slippage, sl_stop=sl_pct, tp_stop=tp_pct)
        stats = portfolio.stats()
        print("\n--- Backtest Performance ---")
        stats_subset = ['Start', 'End', 'Period', 'Start Value', 'End Value', 'Total Return [%]', 'Benchmark Return [%]', 'Max Drawdown [%]', 'Max Drawdown Duration', 'Sharpe Ratio', 'Sortino Ratio', 'Calmar Ratio', 'Total Trades', 'Win Rate [%]', 'Profit Factor', 'Expectancy [%]']
        stats_display = {k: stats.get(k, 'N/A') for k in stats_subset}; print(pd.Series(stats_display))
        if portfolio.trades.count() > 0: fig = portfolio.plot(settings=dict(bm_returns=True)); fig.show()
        else: print("No trades executed, skipping plot.")
        if portfolio.trades.count() > 0:
            drawdown = portfolio.drawdown(); max_dd = drawdown.min() * 100; print(f"\nMax Drawdown: {max_dd:.2f}%")
            returns = portfolio.returns(); var_95 = returns.quantile(0.05) * 100; print(f"95% Value-at-Risk (VaR): {var_95:.2f}% per period")
        else: print("\nNo trades executed.")
        print("--- Backtest Finished ---"); return portfolio, stats
    except Exception as e: print(f"Error during backtesting: {e}"); import traceback; traceback.print_exc(); return None, None


# --- Main Pipeline (Uses Asset-Specific Config) ---
def run_strategy_pipeline(ticker, interval=DEFAULT_INTERVAL, period=DEFAULT_PERIOD):
    print(f"\n{'='*30} Running Pipeline for {ticker} ({interval}) {'='*30}")

    # 1. Data
    data = create_dataframe(ticker, period=period, interval=interval)
    if data is None or data.empty: print("Pipeline stopped: Data fetching failed."); return

    # 2. Features
    data = add_features(data, ticker)
    if data.empty: print("Pipeline stopped: No data after adding features."); return

    # 3. Target
    data = add_target_rolling(data)
    if data.empty: print("Pipeline stopped: Target variable creation failed."); return

    # 4. Split
    X_train, X_val, X_test, y_train, y_val, y_test, test_data_full, feature_cols = \
        train_val_test_split_chronological(data, val_size=0.15, test_size=0.15)
    if X_train.empty or X_val.empty or X_test.empty: print("Pipeline stopped: Split resulted in empty set."); return

    # 5. Model Training w/ Asset-Specific Params
    print("\n--- Model Training ---")
    # Get base LGBM params from asset config or use defaults
    base_lgbm_params = get_config(ticker, 'lgbm_params', {})
    # Define default params that might not be in asset config
    default_lgbm_dict = {
        'objective': 'binary', 'metric': 'binary_logloss',
        'n_estimators': DEFAULT_N_ESTIMATORS, 'learning_rate': 0.01, # Default LR if not specified
        'num_leaves': 50, 'reg_alpha': 0.1, 'reg_lambda': 0.1, # Default structure/reg if not specified
        'random_state': 42, 'n_jobs': -1,
        'class_weight': 'balanced' if DEFAULT_APPLY_CLASS_WEIGHT else None
    }
    # Merge defaults with asset-specific overrides
    lgbm_params = {**default_lgbm_dict, **base_lgbm_params} # Asset params override defaults

    model = lgb.LGBMClassifier(**lgbm_params)
    print(f"Training model for {ticker} with resolved params: {lgbm_params}")
    print(f"Using Early Stopping Rounds: {DEFAULT_EARLY_STOPPING_ROUNDS}")

    eval_metric = 'logloss'
    if len(np.unique(y_val)) < 2: eval_metric = 'accuracy'; print("Warning: Validation set has only one class. Using accuracy for early stopping.")

    try:
        model.fit(X_train, y_train,
                  eval_set=[(X_val, y_val)],
                  eval_metric=eval_metric,
                  callbacks=[lgb.early_stopping(stopping_rounds=DEFAULT_EARLY_STOPPING_ROUNDS, verbose=True),
                             lgb.log_evaluation(period=50)])
        best_iter = model.best_iteration_ if model.best_iteration_ is not None and model.best_iteration_ > 0 else -1
        print(f"Model training complete for {ticker}. Best iteration: {best_iter}")
        if best_iter > 0: model.n_estimators = best_iter # Use actual estimators found

    except Exception as e:
         print(f"Error during model training for {ticker}: {e}. Falling back to basic fit.")
         lgbm_params['n_estimators'] = 100 # Reduce estimators for fallback
         model = lgb.LGBMClassifier(**lgbm_params)
         model.fit(X_train, y_train)
         print("Fallback model training complete."); model.best_iteration_ = -1 # Indicate fallback

    # 6. Optimize Probability Threshold (Use asset-specific target)
    min_precision = get_config(ticker, 'min_precision_target', 0.60) # Get asset specific precision target
    optimized_threshold = find_optimal_threshold(model, X_val, y_val,
                                                metric=DEFAULT_OPTIMIZE_THRESHOLD_METRIC,
                                                min_precision=min_precision)
    if optimized_threshold is None:
        print(f"Stopping pipeline for {ticker}: Could not find suitable threshold.")
        if model is not None and hasattr(model, 'predict_proba'):
             plot_shap_and_importance(model, X_train, X_test, feature_cols, title_suffix=f'for {ticker} ({interval}) (Test Set - No Backtest)')
        return

    # 7. Model Evaluation
    print("\n--- Model Evaluation (Test Set) ---")
    try:
         y_pred_proba_test = model.predict_proba(X_test)[:, 1]
         y_pred_test_optimized = (y_pred_proba_test >= optimized_threshold).astype(int)
         accuracy_test = accuracy_score(y_test, y_pred_test_optimized); precision_test = precision_score(y_test, y_pred_test_optimized, zero_division=0)
         recall_test = recall_score(y_test, y_pred_test_optimized, zero_division=0); f1_test = f1_score(y_test, y_pred_test_optimized, zero_division=0)
         print(f"Test Set Metrics {ticker} (Thresh={optimized_threshold:.4f}): Acc={accuracy_test:.4f}, Prec={precision_test:.4f}, Rec={recall_test:.4f}, F1={f1_test:.4f}")
         plot_confusion_matrix_and_metrics(y_test, y_pred_test_optimized, y_pred_proba_test, title=f'Model Perf {ticker} ({interval}) - Test (Thresh={optimized_threshold:.3f})')
         plot_shap_and_importance(model, X_train, X_test, feature_cols, title_suffix=f'for {ticker} ({interval}) (Test Set)')
    except Exception as e: print(f"Error during test set evaluation or plotting for {ticker}: {e}")

    # 8. Prepare Data for Backtesting
    try:
        test_data_full['pred_proba'] = y_pred_proba_test
    except NameError:
        print("Warning: y_pred_proba_test not available, cannot add to test data.")
        return # Cannot proceed to backtest without probabilities

    if 'atr' not in test_data_full.columns: print("Error: 'atr' column missing."); return
    # Check for ema_trend only if the specific asset config requires it
    use_trend_filter = get_config(ticker, 'use_trend_filter', True)
    if use_trend_filter and 'ema_trend' not in test_data_full.columns: print("Error: 'ema_trend' column missing."); return

    # 9. Backtesting (Pass Ticker)
    run_backtest(
        df_test=test_data_full, ticker=ticker, # Pass ticker here
        init_cash=DEFAULT_INIT_CASH, size=DEFAULT_RISK_PER_TRADE_PERCENT, size_type="percent",
        fees=DEFAULT_FEES, slippage=DEFAULT_SLIPPAGE,
        sl_atr_mult=DEFAULT_SL_ATR_MULTIPLIER, tp_atr_mult=DEFAULT_TP_ATR_MULTIPLIER,
        optimized_proba_threshold=optimized_threshold, interval_str=interval)

    print(f"\n{'='*30} Pipeline Finished for {ticker} ({interval}) {'='*30}")


# --- Main Execution ---
if __name__ == "__main__":
    # Define tickers to run
    tickers_to_run = ["ETH-USD", "BTC-USD", "XRP-USD"]
    chosen_interval = DEFAULT_INTERVAL

    for ticker in tickers_to_run:
        run_strategy_pipeline(ticker, interval=chosen_interval)

    print("\nAll specified tickers processed.")
```
